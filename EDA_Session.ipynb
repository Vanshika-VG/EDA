```python
# We want to divide the cars into 3 categories
k = 3 # Low-Budget, Mid-Range and Luxury

# Initialize centroids to min, median, and max of the MSRP, instead of random
centroids = np.array([df['MSRP'].min(), df['MSRP'].median(), df['MSRP'].max()])

df['Category'] = -1  # indicating unassigned

max_iterations = 20

for iter in range(max_iterations):
    # Assignment step
    for i in range(len(df)):
        distances = np.abs(centroids - df.iloc[i]['MSRP']) # 1D distance (Broadcasting)

        closest_centroid = np.argmin(distances) # Closest Centroid

        if df.at[i, 'Category'] != closest_centroid:
            df.at[i, 'Category'] = closest_centroid


    # Update step
    for i in range(k):  # Now iterating from 0 to k-1
        centroids[i] = df[df['Category'] == i]['MSRP'].mean()

    if iter % 2 == 0:
        print(iter, centroids)

# Map numerical categories (0-indexed) to descriptive labels
category_labels = {0: 'Low-Budget', 1: 'Mid-Range', 2: 'Luxury'}
df['Category'] = df['Category'].map(category_labels)
```

### Outlier Removal

---
#### Z-Score method

> _**Z-score**: also known as standard score or z-value, is a statistical measure that quantifies how many standard deviations a data point is from the mean of a group of data._

      In the context of outlier detection, the Z-score method is commonly used to identify data points that are
      significantly different from the mean of a distribution. The basic idea is that most of the data points in
      a normal distribution lie within a certain number of standard deviations from the mean, and values that
      deviate significantly from this range may be considered outliers.

      The formula for calculating the Z-score of a data point X in a dataset with mean μ and standard deviation
      σ is given by:
$$ Z = (X - μ) / σ $$

      The outliers are the data points that are in the tails of the distribution and therefore far from
      the mean. How far depends on a set threshold Z_thr for the normalized data points zi calculated
      with the formula:

​$$ |Z\_score| > Z\_thr $$

     Commonly used Z_thr values are 2.5, 3.0 and 3.5. Here we will be using 3.0

---

```python
# Perform z-score normalization to detect outliers
from scipy.stats import zscore

df_copy = df.copy()

z_scores = zscore(df_copy[numeric_features])
abs_z_scores = abs(z_scores)

filtered_entries = (abs_z_scores < 3).all(axis=1)
df_filtered = df_copy[filtered_entries]
```

```python
# Visualize the filtered data after removing outliers
plt.figure(figsize=(15, 8))
for i, feature in enumerate(numeric_features, 1):
    plt.subplot(2, 4, i)
    sns.boxplot(x=df_filtered[feature])
    plt.title(f'Boxplot of {feature} (Outliers Removed)')

plt.tight_layout()
plt.show()
```

## 4) Missing Data Analysis:
- Checking the missing values in the raw data
- Discuss the strategies to deal with missing values: imputation, removal, etc.

```python
df = pd.read_csv("data.csv")
```

```python
df.info()
```

```python
# Checking for missing values
missing_values = df.isnull().sum()
print("Missing Values:")
print(missing_values)
print("\n")

# Displaying the raw data
print("Raw Data:")
print(df.loc[538:548, "Engine HP"])
```

```python
# Discussing other strategies to deal with missing values
# 1. Removal of rows with missing values
df_removed = df.dropna()
print("Data after removing rows with missing values:")
print(df_removed.loc[538:548, "Engine HP"])
print("\n")

# 2. Imputation using mean (or median/mode)
numeric_cols = df.select_dtypes(include=np.number).columns
df_imputed = df.fillna(df[numeric_cols].mean())
print("Imputed Data (using mean):")
print(df_imputed.loc[538:548, "Engine HP"])
print("\n")

# 3. Custom Imputation
df_custom_imputed = df.fillna({'Make': 'Unknown', 'Engine HP': 420, 'Engine Fuel Type': 0})
print("Custom Imputed Data:")
print(df_custom_imputed.loc[538:548, "Engine HP"])
print("\n")

# 4. Forward fill or padding(bfill can also be used)
df_forward_fill = df.fillna(method='ffill')
print("Data after forward fill:")
print(df_forward_fill.loc[538:548, "Engine HP"])
print("\n")

# 5. Interpolation (order for interpolation can be changed)
df_interpolated = df.interpolate(method='polynomial', order=2)  # Change the order as needed
print("Data after interpolation:")
print(df_interpolated.loc[538:548, "Engine HP"])
print("\n")
```

## 5) Feature Engineering:

- Creating new features, transformations, and encoding categorical variables.

**Types of features**
- **Numeric**: having a continuous range of values. Can be used directly
- **Categorical**: having values from a finite set of labels.
  - **Nominal**: no ranking or order in labels
  - **Ordinal**: inherent ranking or order in labels


**Feature Encoding**
- **One hot encoding**: each label gets its own feature that can be 0 or 1.  
Eg: Engine Fuel Type  
Use for nominal features.  

- **Label encoding**: each label is assigned a value in some order.  
Eg: assigning size labels (small, medium, large) increasing numbers  
Use for ordinal features  

- **Frequency encoding**: each label is replaced with its frequency  
Eg: replacing city name with its population   
Use to avoid one hot encoding for large number of labels  

- **Mean encoding**: each label is replaced with the mean of its target values  
Eg: replacing course name with average marks received  
Also known as target encoding

```python
def one_hot_encode(df, category):
  one_hot_encoded = pd.get_dummies(df[category], prefix=category)
  df = df.drop(columns = category)
  return pd.concat([df, one_hot_encoded], axis=1)

df = one_hot_encode(df, 'Engine Fuel Type')
```

**Feature Creation**
- **Combination** of multiple features
- **Transformation** of features (eg: log, square)


```python
# Combining features to create new ones
df['Combined MPG'] = df['city mpg'] + df['highway MPG']  # Combines city and highway mileage
df['Price per HP'] = df['MSRP'] / df['Engine HP']       # Price per unit of horsepower


# square transform of existing column "Engine HP
df['Engine HP Squared'] = df['Engine HP'] ** 2            # Calculates the square of engine horsepower

# log transform of "MSRP" (Make sure MSRP doesn't have 0 values)
df['Log MSRP'] = np.log1p(df['MSRP'])  # Use log1p to handle potential 0 values in 'MSRP'

# For adding new columns in numpy
# x_train = np.column_stack((x_train, new_column))
```
